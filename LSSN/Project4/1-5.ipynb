```r
#alpha-1
s = list("Health Care"=0, "Industrials"=0, "Consumer Discretionary"=0,
         "Information Technology"=0, "Consumer Staples"=0, "Utilities"=0,
         "Financials"=0, "Real Estate"=0, "Materials"=0,
         "Energy"=0, "Telecommunication Services"=0)
#all_sectors = readLines("finance_data/Name_sector.csv")
for (sector in sectors){
  s[[sector]] = s[[sector]] + 1
}
```

```r
probs1 = c()
probs2 = c()
for (node in V(g)){
    sector_of_this_node = all_sectors[[node]]
    n_i = length(neighbors(g, node))
    q_i = 0
    #print(all_sectors[[node]])
    for (n in neighbors(g, node))
    {
        if(all_sectors[[n]] == all_sectors[[node]])
            {
            
            q_i = q_i +1
            }
        
    }
    
    probs1 = c(probs1 , (q_i / n_i))
    probs2 = c(probs2 , (s[[sector_of_this_node]] / length(V(g))))
    #print(n_i)
}
```

```r

alpha1 = sum(probs1)/ length(V(g))
alpha2 = sum(probs2)/ length(V(g))
```
