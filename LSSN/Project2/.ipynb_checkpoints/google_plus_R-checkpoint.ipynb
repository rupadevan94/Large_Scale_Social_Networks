```r
library("igraph")
```

```r
files = list.files(path = "./gplus/gplus", pattern = "*.circles")
path = "./gplus/gplus/"
```

```r
numberOfCircles = function(filepath) {
  f = file(filepath, "r")
  count = 0
  while ( TRUE ) {
    line = readLines(f, n = 1)
    if ( length(line) == 0 ) {
      break
    }
    count = count + 1
  }
  close(f)
 return (count)
}
```

```r
file_2circles = list()
for (fname in files) {
    if (numberOfCircles(paste(path,fname,sep = "")) >= 2)
        {
         file_2circles = append(file_2circles, strsplit(fname, "\\.")[[1]][[1]])
    }
}
```

```r
length(file_2circles)
```

```r
# Creating graphs
```

```r
getNodes = function(filepath, egoNode) {
    nodelist = list()
    f = file(filepath, "r")
    while ( TRUE ) {
    line = readLines(f, n = 1)
    if ( length(line) == 0 ) {
      break
    }
    s = strsplit(line, " ")
    nodelist = append(nodelist, c(egoNode, s[[1]][1]))
    nodelist = append(nodelist, c(egoNode, s[[1]][2]))
  }
  close(f)
  return (nodelist)
}
```

```r
graphs = list()
for (f in file_2circles) {
    mat <- scan(paste(path,f,".edges",sep = ""), what="character")
    edge_mat <- matrix(mat, ncol = 2, byrow = TRUE)
    nodes = matrix(unique(mat), ncol=1)
    ego_list = matrix(rep(f, length(nodes)), ncol=1)
    ego_edges = cbind(ego_list, nodes)
    edges = rbind(ego_edges, edge_mat)
    g = graph_from_edgelist(edges, directed=TRUE)
    graphs = append(graphs, list(g))
}
```

```r
ids = c("109327480479767108490", "115625564993990145546", "101373961279443806744")
```

```r
length(graphs)
```

```r
i1 = 0
i2 = 0 
i3 = 0

for (i in 1:length(file_2circles)){
   if (file_2circles[i] == ids[1]){
       i1 = i
   }
   if (file_2circles[i] == ids[2]){
       i2 = i
   } 
    if (file_2circles[i] == ids[3]){
       i3 = i
   } 
}
```

```r
g1 = graphs[[i1]]
g2 = graphs[[i2]]
g3 = graphs[[i3]]
```

```r
dd1_in = degree_distribution(g1, mode = "in")
dd1_out = degree_distribution(g1, mode = "out")

dd2_in = degree_distribution(g2, mode = "in")
dd2_out = degree_distribution(g2, mode = "out")

dd3_in = degree_distribution(g3, mode = "in")
dd3_out = degree_distribution(g3, mode = "out")
```

```r
lab1_in = seq(0, length(dd1_in)-1)
lab1_out = seq(0, length(dd1_out)-1)

lab2_in = seq(0, length(dd2_in)-1)
lab2_out = seq(0, length(dd2_out)-1)

lab3_in = seq(0, length(dd3_in)-1)
lab3_out = seq(0, length(dd3_out)-1)
```

```r
plot(lab1_in, dd1_in, type="l", main="Degree Distribution ID = 109327480479767108490", xlab="Degree", ylab="Number of Vertices")

plot(lab2_in, dd2_in, type="l", main="Degree Distribution ID = 115625564993990145546", xlab="Degree", ylab="Number of Vertices")

plot(lab3_in, dd3_in, type="l", main="Degree Distribution ID = 101373961279443806744", xlab="Degree", ylab="Number of Vertices")
```

```r
plot(lab1_out, dd1_out, type="l", main="Out Degree Distribution ID = 109327480479767108490", xlab="Out Degree", ylab="Number of Vertices")

plot(lab2_out, dd2_out, type="l", main="Out Degree Distribution ID = 115625564993990145546", xlab="Out Degree", ylab="Number of Vertices")

plot(lab3_out, dd3_out, type="l", main="Out Degree Distribution ID = 101373961279443806744", xlab="Out Degree", ylab="Number of Vertices")
```

```r
plot(tail(lab1_out, -1), tail(dd1_out, -1), type="l", main="Out Degree Distribution ID = 109327480479767108490", xlab="Out Degree", ylab="Number of Vertices")

plot(tail(lab2_out, -1), tail(dd2_out, -1), type="l", main="Out Degree Distribution ID = 115625564993990145546", xlab="Out Degree", ylab="Number of Vertices")

plot(tail(lab3_out, -1), tail(dd3_out, -1), type="l", main="Out Degree Distribution ID = 101373961279443806744", xlab="Out Degree", ylab="Number of Vertices")
```
