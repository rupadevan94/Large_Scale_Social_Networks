```r
edgelist_file_path <- "/Applications/LSSCN/P2/facebook_combined.txt"
facebook_network <- read.graph(edgelist_file_path,directed = FALSE)
```

```r
#Question1
is.connected(facebook_network)
```

```r
#Question2
diameter(facebook_network, directed = FALSE)
```

```r
#Question3
deg_distribution<-degree.distribution(facebook_network)
deg_distribution
```

```r
options(repr.plot.width=5,repr.plot.height=5)
plot(deg_distribution,main="Degree distribution ",xlab="Degree",ylab="Frequency")
png("deg_distribution.png")
dev.off()
```

```r
v<-degree(facebook_network)
mean(v)
```

```r
#Question4
degree_log = log(c(1:length(degree.distribution(facebook_network))))
distribution_log = log(degree.distribution(facebook_network))
x_val = degree_log[-which(is.infinite(distribution_log))]
y_val = distribution_log[-which(is.infinite(distribution_log))]
plot(x_val, y_val, main="Degree Distribution in Log Scale",xlab="Degree(log)",ylab="Frequency(log)")
abline(var<-lm(y_val~x_val), col="blue")
var
```

```r
#Question5,6,7
options(repr.plot.width=6,repr.plot.height=6)
personal_nw = induced.subgraph(facebook_network, c(1, neighbors(facebook_network,1)))
vec_size = rep(3.0,vcount(personal_nw))   
vec_size[1]=5       
vec_color = rep("blue",vcount(personal_nw)) 
vec_color[1] ="red"
plot.igraph(personal_nw,vertex.size=vec_size,vertex.label =NA,vertex.color=vec_color)
count_nodes = vcount(personal_nw)
cat("Number of nodes is", count_nodes)
count_edges = ecount(personal_nw)
cat("\nNumber of edges is", count_edges)
diameter_personal_nw= diameter(personal_nw)
cat("\nDiameter of personal network is", diameter_personal_nw)
```
