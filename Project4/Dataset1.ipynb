```python
data = pd.read_csv("network_backup_dataset.csv");
print(data.columns)
X = data[['Week #', 'Day of Week', 'Backup Start Time - Hour of Day','Work-Flow-ID', 'File Name']].values
y = data['Size of Backup (GB)'].values
```

```python
encode_col = [0, 1, 2, 3, 4]
encoders = {}
for col in encode_col:
    le = preprocessing.LabelEncoder()
    le.fit(X[:,col])
    X[:,col] = le.transform(X[:,col])
    encoders[col] = le
```

```python
model = LinearRegression()
model.fit(X,y)
```

```python
model.score(X, y)
```

```python
def get_rmse_folds(model_cv,X,y):
    models  = []
    kf = KFold(n_splits=10)

    rmse = []
    n = 0
    for train, test in kf.split(X,y):
        print("Split: ", n)
        n += 1
        model_cv.fit(X[train],y[train])
        y_pred_train = model_cv.predict(X[train])
        y_pred_test = model_cv.predict(X[test])
        #RMSE Error
        rmse_train = sqrt(mean_squared_error(y[train],y_pred_train))
        rmse_test = sqrt(mean_squared_error(y[test],y_pred_test))
        rmse.append([rmse_train, rmse_test])
        models.append(model_cv)
        
    for i in range(10):
        print ('Fold ' , i, ' Train RMSE: ', rmse[i][0], ' Test RMSE: ', rmse[i][1])
    rmse = np.array(rmse)
    
    plt.plot([i for i in range(10)],rmse[:,0])
    plt.title("RMSE Train for 10 Folds")
    plt.xlabel("Fold")
    plt.ylabel("RMSE Train")
    plt.show()
    
    plt.plot([i for i in range(10)],rmse[:,1])
    plt.title("RMSE Test for 10 Folds")
    plt.xlabel("Fold")
    plt.ylabel("RMSE Test")
    plt.show()
    return models,rmse
```

```python
def plot_graphs(model, X, y, fit = True):
    if (fit):
        model.fit(X,y)
    y_pred = model.predict(X)
    residuals = y - y_pred
    plt.scatter(y, y_pred, s = 0.5)
    plt.xlabel("True Values")
    plt.ylabel("Predicted Values")
    plt.title("Predicted vs True Values")
    plt.show()
    plt.scatter(y_pred, residuals, s = 0.5, color = 'green')
    plt.xlabel("Predicted Values")
    plt.ylabel("Residues")
    plt.title("Residues vs Predicted Values")
    plt.show()
```

```python
model = LinearRegression()
models, rmse = get_rmse_folds(model,X,y)
```

```python
best_model = models[(np.argmin(rmse[:,1]))]
plot_graphs(best_model, X, y, fit = False)
```

```python
# Group Data by Workflow ID
data_wf = pd.read_csv("network_backup_dataset.csv");
print(data_wf.columns)
```

```python
work_ids = (data_wf['Work-Flow-ID'].unique())
data_wf_gp = {}
X_wf = {}
y_wf = {}

for wid in work_ids:
    data_wf_gp[wid] = data_wf.loc[data_wf['Work-Flow-ID'] == wid]
    X_wf[wid] = data_wf_gp[wid][['Week #', 'Day of Week', 'Backup Start Time - Hour of Day', 'File Name']].values
    y_wf[wid] = data_wf_gp[wid]['Size of Backup (GB)'].values
```

```python
encode_col = [0, 1, 2, 3]
encoders_wf = {}
for wid in work_ids:
    encoders_wf[wid] = {}
    for col in encode_col:
        le = preprocessing.LabelEncoder()
        le.fit(X_wf[wid][:,col])
        X_wf[wid][:,col] = le.transform(X_wf[wid][:,col])
        encoders_wf[wid][col] = le
```

```python
for wid in work_ids:
    model = LinearRegression()
    model.fit(X_wf[wid],y_wf[wid])
    
```

```python
def get_rmse_poly(deg, plot = False):
    rmse_all = {}
    avg_rmse_all = []
    for wid in work_ids:
        model = LinearRegression()
        X = X_wf[wid]
        y = y_wf[wid]
        
        if (deg != 1):
            poly = PolynomialFeatures(deg)
            X = poly.fit_transform(X_wf[wid])
        
        kf = KFold(n_splits=10)
        avg_test_rmse = 0.0
        avg_train_rmse = 0.0
        
        rmse = []
        n = 0
        for train, test in kf.split(X,y):
            n += 1
            if (plot):
                print("Split: ", n)
            model.fit(X[train],y[train])
            y_pred_train = model.predict(X[train])
            y_pred_test = model.predict(X[test])
            #RMSE Error
            rmse_train = sqrt(mean_squared_error(y[train],y_pred_train))
            rmse_test = sqrt(mean_squared_error(y[test],y_pred_test))
            rmse.append([rmse_train, rmse_test])
            
            avg_train_rmse += mean_squared_error(y[train],y_pred_train)
            avg_test_rmse += mean_squared_error(y[test],y_pred_test)

        avg_train_rmse = sqrt(avg_train_rmse/10.0)
        avg_test_rmse = sqrt(avg_test_rmse/10.0)
        if (plot):
            print ("Workload: " , wid, "  Average Train RMSE: ", avg_train_rmse)
            print ("Workload: " , wid, "  Average Test RMSE:  ", avg_test_rmse)
        avg_rmse = [avg_train_rmse, avg_test_rmse]
        
        rmse_all[wid] = np.array(rmse)
        avg_rmse_all.append(avg_rmse)
    
    if (plot):
        for wid in work_ids:
            plt.plot([j for j in range(1,11)], rmse_all[wid][:,0], label=wid)
        plt.title("Workflow - RMSE Train for 10 Folds:")
        plt.xlabel("Fold")
        plt.ylabel("RMSE Train")
        plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
        plt.show()

        for wid in work_ids:
            plt.plot([j for j in range(1,11)], rmse_all[wid][:,1], label=wid)
        plt.title("Workflow - RMSE Test for 10 Folds:")
        plt.xlabel("Fold")
        plt.ylabel("RMSE Train")
        plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
        plt.show()
    
    return (avg_rmse_all)
        
```
